A análise das informações sobre o suicídio no Brasil revela a gravidade desse problema de saúde pública tanto no país quanto globalmente. Os dados apresentados indicam que mais de 12 milhões de pessoas tiram suas próprias vidas anualmente no Brasil, colocando-o como o oitavo país com maior número de suicídios no mundo. Esses números são alarmantes e superam a taxa de vítimas de doenças como a AIDS e a maioria dos tipos de câncer.
Diante dessa realidade preocupante, a campanha do Setembro Amarelo foi criada em 2015 como uma resposta para conscientizar as pessoas sobre o suicídio e prevenir sua ocorrência. Embora o Dia Mundial de Prevenção ao Suicídio seja oficialmente celebrado em 10 de setembro, durante todo o mês são promovidos debates, campanhas e ações para aumentar a conscientização sobre o tema.
É preocupante notar que o suicídio é a segunda principal causa de morte entre jovens de 15 a 29 anos. Isso ressalta a importância de implementar políticas eficazes de prevenção e desenvolver estratégias contínuas para reduzir esses números alarmantes e salvar vidas.


Um aspecto importante na análise das taxas de suicídio no Brasil é considerar a relação entre o número de suicídios por 100 mil habitantes ao longo do tempo. Essa métrica é preferível para evitar conclusões errôneas ou distorcidas devido ao crescimento populacional ocorrido entre 1985 e 2016.
Ao observar a tendência das taxas de suicídio no Brasil em relação à taxa mundial, é possível identificar se o país apresenta um comportamento similar ou divergente em relação aos outros países. Comparar essas taxas normalizadas por 100 mil habitantes ajuda a entender se o Brasil está acima ou abaixo da média global em termos de prevalência de suicídios.
Portanto, ao analisar os dados disponíveis, é recomendável considerar a taxa de suicídio por 100 mil habitantes tanto para o Brasil quanto para o contexto mundial. Isso proporciona uma compreensão mais precisa e comparativa das tendências suicidas ao longo do tempo. Essa abordagem leva em conta o tamanho da população e permite uma avaliação mais precisa das mudanças nas taxas de suicídio, independente do crescimento populacional

```python
# criar uma tabela dinâmica
table = pd.pivot_table(df_brasil, values='suicides_no', index=['year'], columns=['age'])

# reordenar as tableas para deixar em ordem crescente
column_order = ['5-14 years', '15-24 years', '25-34 years', '35-54 years', '55-74 years', '75+ years']
table = table.reindex(column_order, axis=1)

# ver as 5 primeiras entradas da pivot table
table.head()
```

```python

#selecionando os paises com a maior taxa de suicidio
suic_sum = pd.DataFrame(df['suicides_no'].groupby(df['country']).sum())
suic_sum = suic_sum.reset_index().sort_values(by='suicides_no',ascending=False)
most_cont = suic_sum.head(5)
fig = plt.figure(figsize=(10,10))
plt.title('Os 5 países com mais casos de suicídios.')
sns.set(font_scale=2)
sns.barplot(y='suicides_no',x='country',data=most_cont,palette="Blues_d")
plt.xticks(rotation=45)
plt.ylabel('Suicídios');
plt.xlabel('Países')
plt.tight_layout()
```

```python

#paises com menos casos de Suicidio.
suic_sum = pd.DataFrame(df['suicides_no'].groupby(df['country']).sum())
suic_sum = suic_sum.reset_index().sort_values(by='suicides_no',ascending=False)
most_cont = suic_sum.tail(5)
fig = plt.figure(figsize=(10,10))
plt.title('Os 5 países com menos casos de suicídios.')
sns.set(font_scale=2)
sns.barplot(y='suicides_no',x='country',data=most_cont,palette="Blues_d")
plt.xticks(rotation=45)
plt.ylabel('Suicídios');
plt.xlabel('Países')
plt.tight_layout()
```

```python
suic_sum = pd.DataFrame(df['suicides/100k pop'].groupby(df['country']).sum())
suic_sum = suic_sum.reset_index().sort_values(by='suicides/100k pop',ascending=False)
most_cont = suic_sum.head(5)
fig = plt.figure(figsize=(10,10))
plt.title('Os 5 países com mais tiveram casos de suicídios por 100 mil habitantes.')
sns.set(font_scale=2)
sns.barplot(y='suicides/100k pop',x='country',data=most_cont,palette="gray_d")
plt.xticks(rotation=45)
plt.ylabel('Suicídios');
plt.xlabel('Países')
plt.tight_layout()
```
